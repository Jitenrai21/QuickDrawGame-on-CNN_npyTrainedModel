# Visualizing .npy Drawing Data
This notebook will load and visualize samples from multiple .npy files in your dataset directory.

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import os
import random
```

```python
# Set the directory containing .npy files
npy_dir = r"D:\QuickDrawDataset(npyfiles)-Trad"
files = os.listdir(npy_dir)
print(f"Found {len(files)} files.")
```

```python
# Randomly select 3 files for visualization
num_files = 3
selected_files = random.sample(files, num_files)
print("Selected files:", selected_files)
```

```python
# Visualize 5 samples from each selected file
samples_per_file = 5
for fname in selected_files:
    arr = np.load(os.path.join(npy_dir, fname))
    print(f"File: {fname}, shape: {arr.shape}")
    fig, axes = plt.subplots(1, samples_per_file, figsize=(15, 3))
    for i in range(samples_per_file):
        # If array is 3D, select i-th sample; if 2D, show as image
        if arr.ndim == 3:
            axes[i].imshow(arr[i], cmap='gray')
            axes[i].set_title(f"Sample {i+1}")
            axes[i].axis('off')
        elif arr.ndim == 2:
            axes[i].imshow(arr[i].reshape(28,28), cmap='gray')
            axes[i].set_title(f"Sample {i+1}")
            axes[i].axis('off')
    plt.suptitle(f"Samples from {fname}")
    plt.show()
```

```python
# Debugging: Print array shape and sample values for each file
for fname in selected_files:
    arr = np.load(os.path.join(npy_dir, fname))
    print(f"File: {fname}, shape: {arr.shape}, dtype: {arr.dtype}")
    # Print first 3 sample values (if possible)
    if arr.ndim == 3:
        for i in range(min(3, arr.shape[0])):
            print(f"Sample {i} stats: min={arr[i].min()}, max={arr[i].max()}, mean={arr[i].mean()}")
            print(arr[i])
    elif arr.ndim == 2:
        print(f"Array stats: min={arr.min()}, max={arr.max()}, mean={arr.mean()}")
        print(arr)
    else:
        print("Array has unexpected shape.")
```
